```python
import csv

file_path = '/home/data-master/GlobeRawData/nytbostonglobecom_2020-07-21/hit_data.tsv'
file_path = '/home/data-master/GlobeRawData/nytbostonglobecom_2020-'


uniqueUsers = {}

listOfFileDates = ['07-01', '07-02','07-03','07-04','07-05','07-06','07-07','07-08','07-09','07-10','07-11','07-12','07-13','07-14','07-15','07-16','07-17','07-18','07-19','07-20','07-21','07-22','07-23','07-24','07-25','07-26','07-27','07-28','07-29','07-30']
#listOfFileDates = ['07-21']

def read_tsv(file_path, date):
    
    file_path += date
    file_path += "/hit_data.tsv"
    
    with open(file_path, 'r', encoding='ISO-8859-1') as file:
        tsv_reader = csv.reader(file, delimiter='\t')
        
        i = 0 
        for row in tsv_reader:
            if (len(row) < 287) : 
                break
            if "BGC Homepage" in row[46] : 
                continue
            if row[287] not in uniqueUsers : 
                uniqueUsers[row[287]] = [[[row[46]], [date]]]
            else : 
                if uniqueUsers[row[287]][-1][1] != [date] : 
                    uniqueUsers[row[287]].append([[row[46]], [date]])
                else : 
                    uniqueUsers[row[287]][-1][0].append(row[46])
            i+=1
        print (i)
    
for date in listOfFileDates : 
    read_tsv(file_path, date)
```

```python
with open('dict.csv', 'w') as csv_file:  
    writer = csv.writer(csv_file)
    for key, value in uniqueUsers.items():
       writer.writerow([key, value])
```
